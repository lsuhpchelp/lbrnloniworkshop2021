This Jupyter notebook try to help you understand the gradient descent concept by finding the minimum value of a given object function $f(x)=(x-1)^2$ within the range of $[0,2]$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rc
#rc('animation', html='jshtml')
```

```python
# object function
def f(x):
    return (x-1)**2;
```

Try to fill in the one line of the derivative of $f(x)$:

```python
# calculate the derivative of f(x)
def df(x):
    # what is the derivative of f(x) ?
    return 2*(x-1);
```

```python
def gradient_descent(start_at=np.random.uniform(-1,1), 
                     iterations=20, 
                     learning_rate=0.1):
    # plot f(x) 
    #xs = np.linspace(start=-1, stop=1)
    #squares = f(xs)
    #plt.plot(xs, squares, color='grey')
    #plt.xlabel('x')
    #plt.ylabel('y = f(x)')
    
    x_vals, y_vals = [], []
    x = start_at
    for _ in range(iterations):
        gradient = df(x) # derivative of f(x) = x**2
        x += learning_rate * -gradient
        y = f(x)
        x_vals.append(x)
        y_vals.append(y)
        #plt.scatter(x, y, color='grey', edgecolor='black', zorder=3)
    return x_vals, y_vals
```

```python
xv, yv = gradient_descent(start_at = 2.0, iterations = 30, learning_rate=0.2)
#print(xv)
#print(yv)
```

```python
from ipywidgets import interact
interact(gradient_descent, 
         start_at=(0,np.pi,0.5), 
         iterations=(1,20,1), 
         learning_rate=(0.05,0.96,0.05),
         __output_result=False);
```

```python
fig, ax = plt.subplots()

t = np.arange(0,2.0, 0.001)
s = f(t)
l = plt.plot(t, s)

redDot, = plt.plot([0], [f(0)], 'ro')
def animate(i):
    # redDot.set_data(i, np.sin(i))
    redDot.set_data(i, f(i))
    return redDot,

# create animation using the animate() function
myAnimation = animation.FuncAnimation(fig, animate, frames=xv, \
                                      interval=200, blit=False, repeat=False)
```
